```python
import pandas as pd
```

```python
df_students = pd.read_csv('students.csv')
students = df_students.dropna().drop_duplicates() # without Null and duplicates

df_teachers = pd.read_csv('teachers.csv')
teachers = df_teachers.dropna().drop_duplicates()  # without Null and duplicates

df_courses = pd.read_csv('courses.csv')
courses = df_courses.dropna().drop_duplicates() # without Null and duplicates

df_student_course = pd.read_csv('student_course.csv')
student_course = df_student_course.dropna().drop_duplicates()  # without Null and duplicates
```

```python
import matplotlib.pyplot as plt
display(students)
print( '1 ) -')
male,female= list(students['gender'].value_counts())

plt.bar(['Male', 'Female'], [male, female], color='b')
plt.title("Students by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
students['gender'].value_counts()
```

```python
print(' 2) -')
age =list(students['age'])

plt.hist(age, bins=15, color='b', alpha=0.7)
plt.title('Histogram')
plt.show()
```

```python
print(' 3) - ')
display(courses)
course_type= dict(courses['course_name'].value_counts())
print(course_type)
c_keys = [k for k in course_type.keys()]
print(c_keys)
c_values = [v for v in course_type.values()]
print(c_values)

plt.figure(figsize=(10, 6)) # larger the page because the letters of the name course not readble
plt.bar(c_keys,c_values, color='b')
plt.title("Number of courses")
plt.xlabel("courses")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
print("4 ) -")

city_count = students[['city']].value_counts()
city_max = city_count.idxmax()
city_min = city_count.idxmin()
display("City with max students", city_max)
display("City with min students", city_min)
```
