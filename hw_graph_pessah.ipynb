```python
import matplotlib.pyplot as plt

years = [2015, 2016, 2017, 2018, 2019]
height = [110, 115, 120, 125, 132]
plt.plot(years, height)
plt.title("Height of a Child Over the Years")
plt.xlabel("Year")
plt.ylabel("Height (cm)")
plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
temperature = [12, 14, 17, 21, 25]
rain = [80, 60, 55, 40, 20]

fig, ax1 = plt.subplots(1, 2, figsize=(12, 5))

ax1[0].plot(months, temperature, color='r')
ax1[0].set_title("Temperature by Month")
ax1[0].set_xlabel("Months")
ax1[0].set_ylabel("Temperature ")

ax1[1].bar(months, rain, color='b')
ax1[1].set_title("Rain by Month")
ax1[1].set_xlabel("Months")
ax1[1].set_ylabel("Rain (mm)")

plt.tight_layout()
plt.show()
```
